```python
df.dropna(axis=0, how='any', inplace=True)#drop null values
```

```python
from sklearn import model_selection
#split data into training data and test data
x_train, x_test, y_train, y_test = model_selection.train_test_split(df['Text'],df['Score'],random_state=2, stratify=df['Score'])
```

# Feature enginerring

```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
```

```python
count_vect = CountVectorizer(analyzer='word', token_pattern=r'\w{1,}',max_features=20000)

xtrain_count =  count_vect.fit_transform(x_train)
xtest_count =  count_vect.transform(x_test)
```

```python
print(count_vect.get_feature_names())
```

```python
print(xtrain_count.toarray().shape) 
```

```python
tfidf_vect = TfidfVectorizer(analyzer='word', token_pattern=r'\w{1,}',max_features=20000)
xtrain_tfidf =  tfidf_vect.fit_transform(x_train)
xtest_tfidf =  tfidf_vect.transform(x_test)
```

```python
print(xtrain_tfidf.toarray().shape) 
```

```python
# ngram level tf-idf 
tfidf_vect_ngram = TfidfVectorizer(analyzer='word', token_pattern=r'\w{1,}', ngram_range=(2,3),max_features=20000)
xtrain_tfidf_ngram =  tfidf_vect_ngram.fit_transform(x_train)
xtest_tfidf_ngram =  tfidf_vect_ngram.transform(x_test)
```

```python
tfidf_vect_ngram.get_feature_names()
```

# Model building

```python
from sklearn import naive_bayes, metrics
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
def train_model(classifier, feature_vector_train, label, feature_vector_valid, is_neural_net=False):
    # fit the training dataset on the classifier
    classifier.fit(feature_vector_train, label)
    
    # predict the labels on validation dataset
    predictions = classifier.predict(feature_vector_valid)
    
    if is_neural_net:
        predictions = predictions.argmax(axis=-1)
    print(classification_report(predictions,y_test))
    i=range(len(predictions))
    return metrics.accuracy_score(predictions, y_test)
```

```python
# Naive Bayes on Count Vectors
accuracy = train_model(naive_bayes.MultinomialNB(), xtrain_count, y_train, xtest_count)
print ("NB, Count Vectors: ", accuracy)

# Naive Bayes on Word Level TF IDF Vectors
accuracy = train_model(naive_bayes.MultinomialNB(), xtrain_tfidf, y_train, xtest_tfidf)
print ("NB, WordLevel TF-IDF Vectors: ", accuracy)

# Naive Bayes on Ngram Level TF IDF Vectors
accuracy = train_model(naive_bayes.MultinomialNB(), xtrain_tfidf_ngram, y_train, xtest_tfidf_ngram)
print ("NB, N-Gram TF-IDF Vectors: ", accuracy)
```

# Grid search

```python
import pandas as pd
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import precision_score, recall_score, accuracy_score, f1_score,confusion_matrix
from sklearn.utils.class_weight import compute_sample_weight
```

```python
pipeline = Pipeline([
    ('vect', TfidfVectorizer(analyzer='word', token_pattern=r'\w{2,3}', max_features=20000)),
    ('clf', naive_bayes.MultinomialNB())
])

parameters = {

    'vect__stop_words': ('english', None),
    'vect__lowercase': (True, False),
    'vect__binary': (True, False),
    'vect__max_features': (20000,10000),
    'vect__ngram_range': ((1, 1), (1, 2)),
}

if __name__ == "__main__":
    grid_search = GridSearchCV(pipeline, parameters, n_jobs=-1, verbose=1, scoring='accuracy', cv=3)#3 3
    grid_search.fit(x_train, y_train)
    print('Best score: %0.3f' % grid_search.best_score_)
    print('Best parameters set:')
    best_parameters = grid_search.best_estimator_.get_params()
    for param_name in sorted(parameters.keys()):
        print('\t%s: %r' % (param_name, best_parameters[param_name]))
    predictions = grid_search.predict(x_test)
    print('Accuracy:', accuracy_score(y_test, predictions))
    print(classification_report(predictions,y_test))
```

```python
from sklearn.linear_model import LogisticRegression
pipeline = Pipeline([
    ('vect', TfidfVectorizer(analyzer='word', token_pattern=r'\w{2,3}', max_features=20000)),
    ('clf', LogisticRegression(penalty='l2',solver='lbfgs'))
])

parameters = {
    'vect__stop_words': ('english', None),
    'vect__lowercase': (True, False),
    'vect__binary': (True, False),
    'vect__max_features': (20000,10000),
    'vect__ngram_range': ((1, 1), (1, 2)),
}

if __name__ == "__main__":
    grid_search = GridSearchCV(pipeline, parameters, n_jobs=-1, verbose=1, scoring='accuracy', cv=3)
    grid_search.fit(x_train, y_train)
    print('Best score: %0.3f' % grid_search.best_score_)
    print('Best parameters set:')
    best_parameters = grid_search.best_estimator_.get_params()
    for param_name in sorted(parameters.keys()):
        print('\t%s: %r' % (param_name, best_parameters[param_name]))
    predictions = grid_search.predict(x_test)
    print('Accuracy:', accuracy_score(y_test, predictions))
    print(classification_report(predictions,y_test))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
pipeline = Pipeline([
    ('vect', TfidfVectorizer(analyzer='word', token_pattern=r'\w{2,3}', max_features=20000)),
    ('clf', GradientBoostingClassifier(random_state=42,n_estimators=100,learning_rate=0.5,max_depth=3))
])

parameters = {
    'vect__stop_words': ('english', None),
    'vect__lowercase': (True, False),
    'vect__binary': (True, False),
    'vect__max_features': (20000,10000),
    'vect__ngram_range': ((1, 1), (1, 2))
               }
    
best_clf=GridSearchCV(pipeline, parameters, cv=3, n_jobs=-1,verbose=1,scoring='accuracy')
best_clf.fit(x_train, y_train)
pred = best_clf.predict(x_test)
print(best_clf.best_params_,best_clf.best_score_  )
print('Accuracy:', accuracy_score(y_test, pred))
print(classification_report(pred,y_test))
```
